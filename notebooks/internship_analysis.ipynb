```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action="ignore", category=Warning)
warnings.filterwarnings("ignore", module="seaborn")
warnings.filterwarnings("ignore", module="matplotlib")
```

```python
# -----------------------------
# LOAD DATASET
# -----------------------------
df = pd.read_csv("Downloads/Data_set 2 - Copy.csv")

print(df.head())
print(df.info())
```

```python
# -----------------------------
# SETTINGS — HIGH CONTRAST STYLE
# -----------------------------
plt.style.use("dark_background")
sns.set_context("talk")
sns.set_style("ticks")
```

```python
# -----------------------------
# FIND NUMERIC & CATEGORICAL
# -----------------------------
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

print("Numeric Columns:", num_cols)
print("Categorical Columns:", cat_cols)
```

```python
# -----------------------------
# FUNCTION — HIGH/LOW COLOR BARS
# -----------------------------
def high_low_colors(series):
    avg = series.mean()
    return ["red" if v < avg else "lime" for v in series]
```

```python
# -----------------------------
# MATPLOTLIB — HISTOGRAMS
# -----------------------------
for col in num_cols:
    plt.figure(figsize=(8,5))
    plt.hist(df[col].dropna(), bins=20, color="cyan", edgecolor="white")
    plt.title(f"Histogram — {col}")
    plt.show()
```

```python
# -----------------------------
# MATPLOTLIB — BAR (HIGH/LOW COLOR)
# -----------------------------
for col in num_cols[:3]:
    sample = df[col].dropna().head(20)
    plt.figure(figsize=(10,5))
    plt.bar(range(len(sample)), sample, color=high_low_colors(sample))
    plt.title(f"Bar — High/Low Colors — {col}")
    plt.show()
```

```python
# -----------------------------
# MATPLOTLIB — LINE PLOT
# -----------------------------
for col in num_cols[:3]:
    plt.figure(figsize=(10,5))
    plt.plot(df[col].dropna().values[:50], color="yellow", marker="o")
    plt.title(f"Line Plot — {col}")
    plt.show()
```

```python
# -----------------------------
# MATPLOTLIB — SCATTER
# -----------------------------
if len(num_cols) >= 2:
    plt.figure(figsize=(7,6))
    plt.scatter(df[num_cols[0]], df[num_cols[1]], c="magenta")
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.title("Scatter Plot")
    plt.show()
```

```python
# -----------------------------
# SEABORN — HISTPLOT
# -----------------------------
for col in num_cols:
    plt.figure(figsize=(8,5))
    sns.histplot(df[col], kde=True, color="orange")
    plt.title(f"Seaborn Hist — {col}")
    plt.show()
```

```python
# -----------------------------
# SEABORN — BOXPLOT
# -----------------------------
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], color="violet")
    plt.title(f"Boxplot — {col}")
    plt.show()

```

```python
# -----------------------------
# SEABORN — VIOLIN
# -----------------------------
for col in num_cols[:3]:
    plt.figure(figsize=(6,4))
    sns.violinplot(x=df[col], color="turquoise")
    plt.title(f"Violin — {col}")
    plt.show()
```

```python
# -----------------------------
# SEABORN — COUNT PLOT (CATEGORICAL)
# -----------------------------
for col in cat_cols[:3]:
    plt.figure(figsize=(8,5))
    sns.countplot(y=df[col], palette="bright")
    plt.title(f"Count Plot — {col}")
    plt.show()
```

```python
# -----------------------------
# SEABORN — PAIRPLOT
# -----------------------------
if len(num_cols) >= 3:
    sns.pairplot(df[num_cols[:4]], palette="husl")
    plt.show()
```

```python
# -----------------------------
# SEABORN — CORRELATION HEATMAP
# -----------------------------
if len(num_cols) >= 2:
    plt.figure(figsize=(10,8))
    corr = df[num_cols].corr()
    sns.heatmap(
        corr,
        annot=True,
        cmap="coolwarm",
        linewidths=1,
        linecolor="black"
    )
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# -----------------------------
# SEABORN — SCATTER WITH COLOR SCALE
# -----------------------------
if len(num_cols) >= 3:
    plt.figure(figsize=(8,6))
    sns.scatterplot(
        data=df,
        x=num_cols[0],
        y=num_cols[1],
        hue=num_cols[2],
        palette="viridis",
        s=120
    )
    plt.title("Scatter — Color by Value")
    plt.show()
```

```python
# -----------------------------
# SEABORN — REGRESSION PLOT
# -----------------------------
if len(num_cols) >= 2:
    plt.figure(figsize=(8,6))
    sns.regplot(
        data=df,
        x=num_cols[0],
        y=num_cols[1],
        scatter_kws={"color":"white"},
        line_kws={"color":"red"}
    )
    plt.title("Regression Plot")
    plt.show()
```

```python
# ==============================
# TASK 1 — DATA OVERVIEW
# ==============================
print("\n===== TASK 1: DATA OVERVIEW =====")
print(df.info())
print(df.describe(include="all"))
```

```python
# ==============================
# FIND COLUMN TYPES
# ==============================
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns
```

```python
# ==============================
# TASK 2 — GENDER DISTRIBUTION
# ==============================
if "Gender" in df.columns:
    plt.figure(figsize=(6,5))
    sns.countplot(x="Gender", data=df, palette="bright")
    plt.title("Gender Distribution")
    plt.show()

    df["Gender"].value_counts().plot(
        kind="pie",
        autopct="%1.1f%%",
        colors=["cyan","magenta","yellow"]
    )
    plt.title("Gender Pie Chart")
    plt.ylabel("")
    plt.show()
```

```python
# ==============================
# TASK 3 — DESCRIPTIVE STATS + HIST
# ==============================
print("\n===== TASK 3: NUMERIC STATS =====")
for col in num_cols:
    print(f"\nColumn: {col}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Std:", df[col].std())

    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True, color="orange")
    plt.title(f"Distribution — {col}")
    plt.show()
```

```python
# ==============================
# TASK 4 — MOST PREFERRED INVESTMENT
# ==============================
pref_col = None
for c in df.columns:
    if "invest" in c.lower():
        pref_col = c
        break

if pref_col:
    print("\n===== TASK 4: MOST PREFERRED =====")
    vc = df[pref_col].value_counts()

    colors = ["lime" if v == vc.max() else "red" for v in vc]

    vc.plot(kind="bar", color=colors, figsize=(8,5))
    plt.title("Most Preferred Investment Avenue")
    plt.show()
```

```python
# ==============================
# TASK 5 — REASONS FOR INVESTMENT
# ==============================
reason_col = None
for c in df.columns:
    if "reason" in c.lower():
        reason_col = c

if reason_col:
    print("\n===== TASK 5: REASONS =====")
    print(df[reason_col].value_counts().head(10))

    plt.figure(figsize=(8,5))
    sns.countplot(y=reason_col, data=df,
                  order=df[reason_col].value_counts().index,
                  palette="cool")
    plt.title("Reasons for Investment")
    plt.show()
```

```python
# ==============================
# TASK 6 — SAVINGS OBJECTIVES
# ==============================
save_col = None
for c in df.columns:
    if "sav" in c.lower():
        save_col = c

if save_col:
    plt.figure(figsize=(8,5))
    sns.countplot(y=save_col, data=df,
                  order=df[save_col].value_counts().index,
                  palette="viridis")
    plt.title("Savings Objectives")
    plt.show()
```

```python
# ==============================
# TASK 7 — INFO SOURCES
# ==============================
info_col = None
for c in df.columns:
    if "source" in c.lower() or "info" in c.lower():
        info_col = c

if info_col:
    plt.figure(figsize=(8,5))
    sns.countplot(y=info_col, data=df,
                  order=df[info_col].value_counts().index,
                  palette="magma")
    plt.title("Investment Information Sources")
    plt.show()
```

```python
# ==============================
# TASK 8 — INVESTMENT DURATION (FIXED)
# ==============================
dur_col = None
for c in df.columns:
    if "duration" in c.lower():
        dur_col = c

if dur_col:

    # Convert text ranges to numeric years
    duration_map = {
        "Less than 1 year": 0.5,
        "1-3 years": 2,
        "3-5 years": 4,
        "More than 5 years": 6
    }

    df["Duration_numeric"] = df[dur_col].map(duration_map)

    print("\nAverage Duration (years):", df["Duration_numeric"].mean())

    # Boxplot using numeric version
    plt.figure(figsize=(7,4))
    sns.boxplot(x=df["Duration_numeric"], color="cyan")
    plt.title("Investment Duration Spread (Numeric)")
    plt.show()

    # Category count plot (better for report)
    plt.figure(figsize=(7,4))
    sns.countplot(y=df[dur_col], palette="bright")
    plt.title("Investment Duration Categories")
    plt.show()

```

```python
# ==============================
# TASK 9 — EXPECTATIONS
# ==============================
exp_col = None
for c in df.columns:
    if "expect" in c.lower():
        exp_col = c

if exp_col:
    plt.figure(figsize=(8,5))
    sns.countplot(y=exp_col, data=df,
                  order=df[exp_col].value_counts().index,
                  palette="cubehelix")
    plt.title("Investment Expectations")
    plt.show()
```

```python
# ==============================
# TASK 10 — CORRELATION ANALYSIS
# ==============================
if len(num_cols) >= 2:
    plt.figure(figsize=(10,8))
    corr = df[num_cols].corr()

    sns.heatmap(
        corr,
        annot=True,
        cmap="coolwarm",
        linewidths=1,
        linecolor="black"
    )
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# Scatter high/low color
if len(num_cols) >= 2:
    x = num_cols[0]
    y = num_cols[1]

    colors = ["red" if v < df[y].mean() else "lime" for v in df[y]]

    plt.figure(figsize=(7,6))
    plt.scatter(df[x], df[y], c=colors)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title("Correlation Scatter (Low=Red, High=Lime)")
    plt.show()
```
